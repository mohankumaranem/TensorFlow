1) Write a function in python that inputs a dataframe and identify which columns have date in them. Using these date columns make new columns which are difference between these columns taking 2 at a time. (for instance if there is date1, date2, date3 columns, output should be like date1-date2, date2-date3, date1-date3). For this problem only, print out data in the colab.

Thing to consider

· Code should be efficient and fast

· Code should be well commented and easy to interpret

· **here dataset made by me was made in excel hence it the result will be constant output because i used excel short cuts to create dataset and negative values are obtained due to year 1999  subracted from year 2000 and the output are in days and can be conveeted into months or years.**

```
def ConvertDate(df):                #Data Frame Passed To the Function
  df2=pd.DataFrame()                #created A new DataFrame To find all The date Features
  j=df.columns.to_list()            #A list Contains All Column Names of Dataframe
  for i  in j:
    df2[i]=pd.to_datetime(df[i],errors='coerce')                      #we are Copying all Cloumns  Containing Dates to second dataframe
  df2 = df2.dropna(how='all', axis=1)                                 #Drop all Features Which are not dates in Second Data Frame
  k=np.asanyarray(df2.select_dtypes(include=['datetime64']).columns)  # To calculate How many Features Containing Dates
  for i in  range(0,len(k)):        #Loop To Caluclate The Differences between Time
    if i==len(k)-1:
      df2['date'+str(i+1)+'-date'+str(len(k)-i)]=(df2[k[0]]-df2[k[-1]]).dt.days
    else:    
      df2['date'+str(i+1)+'-date'+str(i+2)]=(df2[k[i]]-df2[k[i+1]]).dt.days
  df2.dropna(inplace=True)            #Drop all Nan and NaT values In new DatFrame
  print(df2.iloc[0:,4:])              #To Show All the differences of Dates
                                      # I Used Excel Short Cuts for dataSet hence The Diffecernce Is Constant Between Features
  return None


import numpy as np
import pandas as pd                   #importing all necessary libraries for the code
import datetime  as dt
df=pd.read_excel('/content/Book2.xlsx')        #Using My Own Data Set using Book2.Xlsx
ConvertDate(df)                       #Function for Date Difference
```

2.Write a function in python that take dataframe as input and drop columns having Pearson correlation more than 0.85

Thing to consider

· Code should drop least amount of variable as possible. (this is an important point)

· Code should be efficient and fast

· Code should be well commented and easy to interpret

· Use google Colab

· Code should be robust enough to run on any dataset

· Make a dummy dataset by yourself or pass any publicly available dataset to test out your logic

```
def DropPearson(df3):
  corr_matrix = df3.corr(method='pearson').abs()                                       # Calculationg Pearson Correlaton for the entire Dataframe
  upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))  # Select upper triangle of correlation matrix
  to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]          # Find index of feature columns with correlation greater than 0.85
  df3.drop(df3[to_drop], axis=1,inplace=True)                                          # Drop features 
  return df3

df3=pd.read_csv('/content/module_5_auto.csv')
DropPearson(df3)
df3                                                                                    # Priting DataFrame After Dropping the Features

```
